##Recursion

###Factorial

````
def factorial(n):
  """Return the factorial of positive integer n"""
  if n == 1:
    return 1
  else:
      return n *factorial(n-1)
````

###sum_digits

````
def sum_digits(n):
  """Return the sum of the digits of positive integer n."""
  if n < 10:
    return n
  else:
    all_but_last = n // 10
    last = n % 10
    return sum_digits(all_but_last) + last
````

###Fib

````
def fib(n):
  """Return the nth number in the Fibinacci sequence"""
  if n == 1:
    return 0
  elif n == 2:
    return 1
  else:
    return fib(n -1) + fib(n-2)
````

###isPalindrome

````
def isPalindrome(s):
  """Return whether string s is a palindrome."""
  if len(s) <= 1:
    return True
  else:
    return s[0] == s[-1] and isPalindrome(s[1:-1])
````

````
isPalindrome("mom")
````

````
isPalindrome("dad")
````
